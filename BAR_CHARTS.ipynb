```python
csv_path = "lyrics_output/df_words_All_Out_10s.csv"
df_10 = pd.read_csv(csv_path)
df_10=df_10.sort_values("count", ascending=False).reset_index()
```

```python
df_10
```

```python
df_10["Group"] = ""
```

```python
x=df_10.set_index("Unnamed: 0")
```

```python
y = x.loc["you"]["count"] + x.loc["your"]["count"] + x.loc["you're"]["count"] + x.loc["yours"]["count"] + x.loc["you'd"]["count"]
y
```

```python
z = x.loc["i"]["count"] + x.loc["i'm"]["count"] + x.loc["my"]["count"] + x.loc["mine"]["count"] + x.loc["i'd"]["count"]
z
```

```python
df = pd.DataFrame({"You words": [y], "I words": [z]})
```

```python
df
```

```python
x_axis = ["You Words", "I Words"]
y_axis = [df.loc[0]["You words"], df.loc[0]["I words"]]
```

```python
plt.bar(x_axis, y_axis)
plt.show()
```
