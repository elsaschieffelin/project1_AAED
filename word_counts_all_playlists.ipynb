```python
import spotipy
import pandas as pd
from pprint import pprint
import spotipy.util as util
from spotipy.oauth2 import SpotifyClientCredentials
import requests
import spotipy.oauth2 as oauth2
from bs4 import BeautifulSoup
from api_key import genius_api_key
from api_key import CLIENT_ID
from api_key import CLIENT_SECRET



### create functions

def request_song_info(song_title, artist_name):
    main_url = 'https://api.genius.com/search'
    headers = {'Authorization': 'Bearer ' + genius_api_key}
    data = {'q': song_title + ' ' + artist_name}
    response = requests.get(main_url, data=data, headers=headers)

    return response

def get_song_lyrics(url):
    page = requests.get(url)
    html = BeautifulSoup(page.text, 'html.parser')
    lyrics = html.find('div', class_='lyrics').get_text()

    return lyrics

# def word_count(str):
#     counts = dict()
#     words = str.split()

#     for word in words:
#         if word in counts:
#             counts[word] += 1
#         else:
#             counts[word] = 1

#     return counts

def numList(list):
        dic = {}
        for num in list:
            if num in dic.keys():
                dic[num] += 1
            else:       
                key = num
                value = 1
                dic[key] = value
        return dic

def call_playlists (user_id_str, playlist_id_str, offset_value):
    playlists=sp.user_playlist_tracks(user=user_id_str,playlist_id=playlist_id_str, offset=offset_value, limit=None)
    return playlists

##Set Spotify API tokens
credentials = oauth2.SpotifyClientCredentials(
       client_id=CLIENT_ID,
       client_secret=CLIENT_SECRET)

token = credentials.get_access_token()
spotify = spotipy.Spotify(auth=token)
sp = spotipy.Spotify(auth=token)

```

```python

##Set empty lists to use later
pl=1
gbl = globals()
dataframes=[]
for item in playlist_dict:
    song_titles_list=[]
    artist_names_list=[]
    list_of_ids=[]
    tracks_top=[]
    words_collected=[]
    sep_words=[]
    full_list=[]
    song_titles_list[:] = []
    artist_names_list[:] = []
    list_of_ids[:] = []
    tracks_top[:] =[]
    words_collected[:] = []
    sep_words[:] = []
    full_list[:] = []
    song_titles_list=[]
    artist_names_list=[]
    list_of_ids=[]
    tracks_top=[]
    words_collected=[]
    sep_words=[]
    full_list=[]

    offset_value=0
    x=0
    
    user_id_str=playlist_dict[pl]["user_id_str"]
    playlist_id_str=playlist_dict[pl]["playlist_id_str"]
    playlist_len=playlist_dict[pl]["playlist_len"]
    playlist_title=playlist_dict[pl]["playlist_title"]
    try:
        while len(artist_names_list) <= playlist_len:
            offset_value=0
            playlists=call_playlists(user_id_str, playlist_id_str,offset_value)
            artist_name=playlists["items"][x]['track']['artists'][0]['name']
            song_name=playlists["items"][x]['track']['album']['name']
            song_id=playlists["items"][x]['track']['id']
            if song_id not in list_of_ids:
                artist_names_list.append(artist_name)
                song_titles_list.append(song_name)
                list_of_ids.append(song_id)
                x+=1
                offset_value=len(artist_names_list)
                call_playlists(user_id_str, playlist_id_str, offset_value)
            else:
                x+=1
    except IndexError:
        print(f"You've saved {len(artist_names_list)} songs from this {playlist_title} playlist")

    d = list(zip(song_titles_list,artist_names_list))
    x=0
    tracks_top=[]
    for items in d:
        artist=artist_names_list[x]
        song=song_titles_list[x]
        song_dict={"song_title":song, "artist_name":artist}
        tracks_top.append(song_dict)
        x+=1

    x=0
    words_collected=[]
    for item in tracks_top:
        song_title=tracks_top[x]['song_title'].lower()
        artist_name=tracks_top[x]['artist_name'].lower()
        request_song_info(song_title, artist_name).text

        response = request_song_info(song_title, artist_name)
        data = response.json()
        remote_song_info = None

        for hit in data['response']['hits']:
            if artist_name.lower() in hit['result']['primary_artist']['name'].lower():
                remote_song_info = hit
                break

        if remote_song_info:
            song_url = remote_song_info['result']['url']

            clean_lyrics=get_song_lyrics(song_url).replace("\n"," ")
            clean_lyrics=clean_lyrics.replace("[","")
            clean_lyrics=clean_lyrics.replace("]","")
            clean_lyrics=clean_lyrics.replace(":","")
            clean_lyrics=clean_lyrics.replace("  "," ")
            clean_lyrics=clean_lyrics.replace("(","")
            clean_lyrics=clean_lyrics.replace(")","")
            clean_lyrics=clean_lyrics.replace("&","")
            clean_lyrics=clean_lyrics.replace(",","")
            clean_lyrics=clean_lyrics.replace("\ ","")
            clean_lyrics=clean_lyrics.replace("\'","'")
            clean_lyrics=clean_lyrics.lower()
            words_collected.append(clean_lyrics)

        x+=1

    y=0
    sep_words=[]
    for item in words_collected:

        sep=item.split()
        sep_words.append(sep)
        y+=1

    full_list=[]
    for lyrics_song in sep_words:
        for item in lyrics_song:
            full_list.append(item)

    full_list=numList(full_list)

    gbl['df_words_'+playlist_title]=pd.DataFrame(full_list, index=["count"]).T
    gbl['df_words_'+playlist_title].sort_values("count", ascending=False)
    pl+=1
```
