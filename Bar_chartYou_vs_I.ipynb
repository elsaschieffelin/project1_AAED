```python
csv_path="lyrics_output/df_words_United_States_Top_50.csv"
df_Top=pd.read_csv(csv_path)
```

```python
x=0
decade_list = [df_00, df_10, df_50, df_60, df_70, df_80, df_90, df_Top]
y = ["00s", "10s", "50s", "60s", "70s", "80s", "90s", "Top 50"]
```

```python

for decade in decade_list: 
    df = decade_list[x]
    name = y[x]
    df=df.set_index("Unnamed: 0")
    try: 
        you = df.loc["you"]["count"] + df.loc["your"]["count"] + df.loc["you're"]["count"] + df.loc["yours"]["count"] + df.loc["you'd"]["count"]
        i = df.loc["i"]["count"] + df.loc["i'm"]["count"] + df.loc["my"]["count"] + df.loc["mine"]["count"] + df.loc["i'd"]["count"]
        x_axis = ["You words", "I words"]
        y_axis = [you, i]
        plt.bar(x_axis, y_axis)
        plt.title(f"Decade {name} You vs. I")
        plt.show()
    except KeyError:
        print(f"no label in {name}")
    except IndexError:
        print(f"no index")
    x+=1
```
