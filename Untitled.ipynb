```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_one = "Output_Files/results.csv" 
file_two = "Output_Files/results_avg.csv" 

file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1") 
file_one_df=file_one_df.set_index("Unnamed: 0")
column_names=list(file_one_df)

file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1") 
file_two_df=file_two_df.set_index("Unnamed: 0")
column_names_1=list(file_two_df)

file_one_df=file_one_df[["All Out 50s", "All Out 60s", "All Out 70s", "All Out 80s", "All Out 90s", "All Out 00s", "All Out 10s", "US Top 50"]]
file_one_df
```

```python
variety_words=pd.DataFrame(file_one_df.count())

variety_words
```

```python
x_axis=list(file_one_df)
y_axis=variety_words.apply(lambda x: x.tolist(), axis=1)
```

```python
plt.bar(x_axis, y_axis, color='r', alpha=0.5, align="center",width=.5)
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis, rotation='vertical')
plt.title("Count of Unique Words by Playlist")
plt.xlabel("Playlists")
plt.ylabel("Count of Words")
plt.savefig("Output_Files/unique_words_plot.png")
plt.show()
```

```python
#Average Unique Words Per Song
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1") 
file_two_df=file_two_df.set_index("Unnamed: 0")
file_two_df=file_two_df.T.sort_values('count', ascending=False)

artists_avg=[]

for items in file_two_df.index.values:
    if "This_Is" in items:
        artists_avg.append(items)
        
#artists_avg
```

```python
all_artists=[]
all_values=[]

for index, row in file_two_df.iterrows():
    if index in artists_avg:
        all_artists.append(index)
        
all_artists
```

```python
import re

df_names_artists=file_two_df.loc[all_artists]
df_names_artists=df_names_artists.reset_index()

names_artist=df_names_artists['index'].values.tolist()

cleaned_index=[]
for items in names_artist:
    cleaned_item=re.sub(r'_', ' ',items)
    cleaned_index.append(cleaned_item)
cleaned_index_1=[]
for items in cleaned_index:
    cleaned_item=re.sub(r'df words This Is ', '',items)
    cleaned_index_1.append(cleaned_item)
      
cleaned_index_1
```

```python
# set width of bar
barWidth = 0.5
 
# set height of bar
x_axis = cleaned_index_1
y_axis = df_names_artists['count'].tolist()

plt.bar(x_axis, y_axis, color='c', alpha=0.5, align="center",width=.5)

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis, rotation='vertical')
plt.title("Count of Unique Words per Song by Artist")
plt.xlabel("Artist")
plt.ylabel("Count of Words")
plt.tight_layout()
plt.grid()
plt.savefig("Output_Files/unique_words_artist_song_plot.png")
plt.show()
```
